```python
import re

def list1(text):
  a=re.findall(r'(?:(\d+)\. (.+?)\n?)',text)
  for i in range(1,len(a)+1):
    if i == 1:
      text =re.sub(r'(?:1\. (.+))',r'<ol>\n<li>\1</li>', text, count=1) # identifica a primeira linha da lista numerada e adiciona o <ol>
    elif i == len(a):
      text =re.sub(rf'(?:{len(a)}\. (.+))',r'<li>\1</li>\n</ol>', text, count=1) # identifica a ultima linha da lista e adiciona o </ol>
    else:
      text = re.sub(rf'(?:{i}\. (.+))',r'<li>\1</li>', text,count=1) # faz a substituição normal da linha correspondente
  return text
  
def list2(text):
    a = ""
    i = 0
    for m in re.finditer(r'((?:\d+\. .+(?:\n|$))+)', text): # identifica as várias listas existentes no texto
        a += text[i:m.start()] # adicionamos todo o texto antes da lista na posição i no array a 
        a += list1(m.group(1)) # fazemos a conversão da lista para html e adicionamos ao fim do array a
        i = m.end() # avançamos o apontador i do array para a possição do fim da lista convertida 
    a += text[i:] # adicionamos todo o texto depois da uktima lista convertida 
    return a

def conv(text):
   er11=re.compile(r'^### (.+)$',re.MULTILINE) # Cabeçalho ###
   text=er11.sub(r'<h3>\1</h3>',text)
   
   er12=re.compile(r'^## (.+)$',re.MULTILINE) # Cabeçalho ##
   text=er12.sub(r'<h2>\1</h2>',text)
   
   er13=re.compile(r'^# (.+)$',re.MULTILINE) # Cabeçalho #
   text=er13.sub(r'<h1>\1</h1>',text)
   
   er2=re.compile(r'\*\*(.+?)\*\*') # Bold
   text=er2.sub(r'<b>\1</b>',text)
   
   er3=re.compile(r'\*(.+?)\*') # Itálico
   text=er3.sub(r'<i>\1</i>',text)
   
   text=list2(text) # Listas
   
   er4=re.compile(r'\!\[(.+?)\]\((.+?)\)') # Imagens
   text=er4.sub(r'<img src="\2" alt="\1"/>',text)
   
   er5=re.compile(r'\[(.+?)\]\((.+?)\)') # Links
   text=er5.sub(r'<a href="\2">\1</a>',text)
  
   return(text)



texto = """# Título Principal
## Subtítulo
Este é um **exemplo em bold** e este é um *exemplo em itálico*.

Lista numerada:
1. Primeiro item
2. Segundo item
3. Terceiro item

Aqui está um [link para a UC](http://www.uc.pt).

Aqui está uma imagem: ![um coelho](http://www.coellho.com)
"""
print(f"----------ORIGINAL----------\n{texto}")
print("---------CONVERTIDO---------")
print(conv(texto))
```
